# Matplotlib 相关概念

## Matplotlib 绘图步骤
（1）创建画布/图纸（figure）

（2）在画布上创建一个或者多个绘图区域

（3）在plotting区域上描绘点，线等

（4）为plotting（绘图线或者坐标轴）添加修饰标签

（5）其他各种DIY（如：旋转标签，添加图例）


## Matplotlib 基本元素

（1）变量（数据）

（2）函数

（3）画布

## 层级
（1） figure 画布

（2） axes 坐标系

（3） x axis， y axis， title， data

（4） tick，label

（5） tick lable

# Matplotlin figure 画布
1.Matplotlib的图像位于figure对象中

2.通过 plt.figure() 创建

1>如果没有指定创建，matplotlib会自动生成默认的画布

2>参数 figsize 控制画布大小，单位为英寸（inch），1 inch = 2.54 cm

3. plt.show() 显示绘图结果

注意，在jupyter notebook中使用matplotlib时，代码块要放在一个cell中

```python
# 引入包
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 创建画布
plt.figure()

# 生成数据
data = [2,4,63,8,10]

# 可视化结果
plt.plot(data)
plt.show()
```

```python
data = np.array([1,2,3,4,5])
plt.plot(data)
plt.show()
```

```python
# 设置画布
plt.figure(figsize = (20,5))

# 生成数据

random_data = np.random.randn(100)

# 画图
plt.plot(random_data)
plt.show()
```

## 可视化优化(绘制子图)

plt.subplots(nrows, ncols, sharex, sharey)

nrows, ncols 分割的行数和列数

sharex, sharey 是否共享x轴、y轴

返回新创建的figure和subplot数组

```python
fig, subplot_arr = plt.subplots(2,2,figsize=(8,8))

subplot_arr[0,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[0,1].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,1].scatter(np.random.randn(50),np.random.randn(50)**2)
```

```python
fig, subplot_arr = plt.subplots(2,2,figsize=(8,8),sharex = True, sharey = True)

subplot_arr[0,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[0,1].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,1].scatter(np.random.randn(50),np.random.randn(50)**2)
```

## 坐标刻度

设置刻度范围

plt.xlim(), plt.ylim()

ax.set_xlim(), ax.set_ylim()

设置显示的刻度

plt.xticks(), plt.yticks()

ax.set_xticks(), ax.set_yticks()

设置刻度标签

如果使用plt设置刻度标签，需要使用 plt.xticks(), plt.yticks()进行配置

ax.set_xticklabels(), ax.set_yticklabels()

设置坐标轴标签

ax.set_xlabel(), ax.set_ylabel()

设置标题

plt.title()

ax.set_title()

图例

绘制图形时，使用label参数

ax.legend(loc), plt.legend(loc)

loc=‘best’自动选择放置图例最佳位置

中文显示

使用rcParams属性进行配置

a.指定默认字体

matplotlib.rcParams['font.sans-serif'] = ['SimHei’]

b.解决保存图像是负号'-'显示为方块的问题

matplotlib.rcParams['axes.unicode_minus'] = False

```python
# 生成数据
data1 = np.random.randn(1000).cumsum()
data2 = np.random.randn(1000).cumsum()
data3 = np.random.randn(1000).cumsum()
data4 = np.random.randn(1000).cumsum()
```

```python
plt.figure()

plt.plot(data1)
plt.plot(data2)
plt.plot(data3)
plt.plot(data4)

plt.show()
```

### 例子1 (plt)

```python
plt.figure()

# 设置刻度范围
plt.xlim([0,500])

# 设置显示的刻度
plt.xticks([0,100,200],['x1','x2','x3'])

# 设置坐标轴标签
plt.xlabel('Number')
plt.ylabel('Random')

# 设置标题
plt.title('Example')

# 图例
plt.plot(data1,label = 'line1')
plt.plot(data2, label = 'line2')
plt.plot(data3, label = 'line3')

plt.legend(loc = 'best')
plt.show()
```

### 例子2 (ax)

```python
fig, ax = plt.subplots(1)
ax.plot(data1, label = 'label 1')
ax.plot(data2, label = 'label 2')
ax.plot(data3, label = 'label 3')

ax.set_xlim([0,800])

ax.set_xticks([0,100,200,300,400,500])

ax.set_xticklabels(['x1','x2','x3','x4','x5'])

ax.set_xlabel('Number')
ax.set_ylabel('Random')

ax.set_title('Example2')

ax.legend(loc = 'best')
plt.show()
```

### 例子3（字体）

```python
import matplotlib

matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(1)
ax.plot(data1, label = 'label 1')
ax.plot(data2, label = 'label 2')
ax.plot(data3, label = 'label 3')

ax.set_xlim([0,800])

ax.set_xticks([0,100,200,300,400,500])

ax.set_xticklabels(['x1','x2','x3','x4','x5'])

ax.set_xlabel('数字')
ax.set_ylabel('随机数')

ax.set_title('Example2')

ax.legend(loc = 'best')
plt.show()
```

## 散点图
散点图

plt.scatter(x, y, s, c, marker)

x 横轴的数据 

y 纵轴的数据

s 点的大小，标量或数组

c 点的颜色，标量或数组

marker 点的样式

其中：标量：同时所有的点的大小（颜色）；数组：为每个点设置不同的大小（颜色）

颜色配置： https://matplotlib.org/api/_as_gen/matplotlib.pyplot.colors.html

点的样式： https://matplotlib.org/api/markers_api.html

```python
plt.figure()

x = np.random.randn(50)
y = x*2

plt.scatter(x,y)
plt.show()
```

```python
plt.figure()

# 生成两组数据

x = np.random.randn(50)
y1 = x ** 2
y2 = x * 2

# 可视化结果
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.show()
```

## 柱状图

plt.bar(x, height, color)

x 柱子的横坐标

height 柱子的高度（数据）

color 柱子的颜色

注意：绘制多组柱状图时，需要设置不同的横坐标，避免重叠

```python
# 单组数据 
x = np.array([1,2,3,4,5])
data = np.array([6,4,2,6,2])
plt.bar(x,data)
plt.show()
```

```python
# 多组数据

x = np.array([1,2,3,4,5])
data = np.array([6,4,2,6,2])
plt.bar(x,data)


x2 = x
data2 = [2,6,2,5,1]
plt.bar(x,data2, color = 'g')
plt.show()
```

**数据被遮盖**

```python
# 多组数据

x = np.array([1,3,5,7,9])
data = np.array([6,4,2,6,2])
plt.bar(x,data)


x2 = x + 1
data2 = [2,6,2,5,1]
plt.bar(x2,data2, color = 'g')
plt.show()
```

```python
x = np.random.randint(1,30,size=10)

data1 = abs(x-np.mean(x))

data2 = x-np.mean(x)

plt.figure()

plt.scatter(x,data1,s=x*3,c="b",marker="*")

plt.scatter(x,data2,s=x[0]*5,c="r",marker="h")

plt.show()
```

```python
plt.figure(figsize=(10,5))

year=["2016","2017","2018","2019"]

x_A=[1,3,5,7]

score_A = [40,50,59,60]

w =0.5

plt.bar(x_A,score_A,width = w,label="A")

 

x_B=np.array(x_A)+w

score_B = [50,70,80,90]

plt.bar(x_B,score_B,width=w,label="B")

plt.xlabel("year")

plt.ylabel("score")

plt.legend()

plt.xticks(x_B-0.25,year)

plt.show()
```

## 直方图（histgram）
直方图形式类似于柱状图 ，通过将数据组合在一起来显示分布，比如：显示各年龄段的数据个数

plt.hist(x, bins, rwidth )

x 数据

rwidth 柱子宽度比例，默认为1

bins 分组的个数或分组边界，如4或[20, 30, 40, 50, 60]

除了最后一组，其余组的右边界是开放的
[20, 30), [30, 40), [40, 50), [50, 60]

```python
data  = np.random.randn(10000)
```

```python
plt.figure()
plt.hist(data, 40)
plt.show()
```

```python
bins = [-4,-3,-2,-1,0,1,2,3,4]
plt.hist(data,bins,rwidth= 0.9)
plt.show()
```

## 矩阵绘图

cmap 颜色主题
https://matplotlib.org/users/colormaps.html

```python
plt.imshow(data, cmap = plt.cm.gnuplot)
plt.colorbar()
plt.show()
```

### ticks 设置

```python
data=np.random.randint(1,100,size=20).reshape(4,5)

plt.imshow(data,cmap=plt.cm.Blues)

plt.colorbar()

plt.xticks(range(data.shape[1]),["2014","2015","2016","2017","2018"])

plt.yticks(range(data.shape[0]),["January","February","March","April"])

plt.xlabel("year")

plt.ylabel("month") 
```

## 折线图
plt.plot(x, y, fmt) 绘制折线图

x, y 横轴纵轴数据

fmt 颜色 标记 线型的样式

如 r.-- 表示 红色 点标记 虚线

```python
fig, subpplot_arr = plt.subplots(2,2,figsize = (8,8))
data = np.random.rand(20)

subpplot_arr[0,0].plot(data, '--r.')
subpplot_arr[0,1].plot(data, 'gv:')
subpplot_arr[1,0].plot(data, 'b<-')
subpplot_arr[1,1].plot(data, 'ys-.')
plt.show()
```

```python
plt.rcParams['font.sans-serif']=['SimHei']

fig,ax = plt.subplots(2,1,figsize=(10,5))

data = np.random.randint(1,11,size=5)

ax[0].plot(data,"r--*")

ax[1].plot(data**2,"bo-")

plt.xlabel("随机数")

plt.show()
```
