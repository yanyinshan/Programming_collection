# Matplotlib 相关概念

## Matplotlib 绘图步骤
（1）创建画布/图纸（figure）

（2）在画布上创建一个或者多个绘图区域

（3）在plotting区域上描绘点，线等

（4）为plotting（绘图线或者坐标轴）添加修饰标签

（5）其他各种DIY（如：旋转标签，添加图例）


## Matplotlib 基本元素

（1）变量（数据）

（2）函数

（3）画布

## 层级
（1） figure 画布

（2） axes 坐标系

（3） x axis， y axis， title， data

（4） tick，label

（5） tick lable

# Matplotlin figure 画布
1.Matplotlib的图像位于figure对象中

2.通过 plt.figure() 创建

1>如果没有指定创建，matplotlib会自动生成默认的画布

2>参数 figsize 控制画布大小，单位为英寸（inch），1 inch = 2.54 cm

3. plt.show() 显示绘图结果

注意，在jupyter notebook中使用matplotlib时，代码块要放在一个cell中

```python
# 引入包
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 创建画布
plt.figure()

# 生成数据
data = [2,4,63,8,10]

# 可视化结果
plt.plot(data)
plt.show()
```

```python
data = np.array([1,2,3,4,5])
plt.plot(data)
plt.show()
```

```python
# 设置画布
plt.figure(figsize = (20,5))

# 生成数据

random_data = np.random.randn(100)

# 画图
plt.plot(random_data)
plt.show()
```

## 可视化优化(绘制子图)

plt.subplots(nrows, ncols, sharex, sharey)

nrows, ncols 分割的行数和列数

sharex, sharey 是否共享x轴、y轴

返回新创建的figure和subplot数组

```python
fig, subplot_arr = plt.subplots(2,2,figsize=(8,8))

subplot_arr[0,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[0,1].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,1].scatter(np.random.randn(50),np.random.randn(50)**2)
```

```python
fig, subplot_arr = plt.subplots(2,2,figsize=(8,8),sharex = True, sharey = True)

subplot_arr[0,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[0,1].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,0].scatter(np.random.randn(50),np.random.randn(50)**2)
subplot_arr[1,1].scatter(np.random.randn(50),np.random.randn(50)**2)
```

# 图种类


```python
plt.figure()

# 生成两组数据

x = np.random.randn(50)
y1 = x ** 2
y2 = x * 2

# 可视化结果
plt.scatter(x, y1, s= 100, c = 'r', marker = 'x')
plt.scatter(x, y2, s = 20, c = 'g', marker = 's')
plt.show()
```

## 散点图
散点图

plt.scatter(x, y, s, c, marker)

x 横轴的数据 

y 纵轴的数据

s 点的大小，标量或数组

c 点的颜色，标量或数组

marker 点的样式

其中：标量：同时所有的点的大小（颜色）；数组：为每个点设置不同的大小（颜色）

颜色配置： https://matplotlib.org/api/_as_gen/matplotlib.pyplot.colors.html

点的样式： https://matplotlib.org/api/markers_api.html

```python
plt.figure()

x = np.random.randn(50)
y = x*2

plt.scatter(x,y)
plt.show()
```

```python
plt.figure()

# 生成两组数据

x = np.random.randn(50)
y1 = x ** 2
y2 = x * 2

# 可视化结果
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.show()
```

## 柱状图

plt.bar(x, height, color)

x 柱子的横坐标

height 柱子的高度（数据）

color 柱子的颜色

注意：绘制多组柱状图时，需要设置不同的横坐标，避免重叠

```python
# 单组数据 
x = np.array([1,2,3,4,5])
data = np.array([6,4,2,6,2])
plt.bar(x,data)
plt.show()
```

```python
# 多组数据

x = np.array([1,2,3,4,5])
data = np.array([6,4,2,6,2])
plt.bar(x,data)


x2 = x
data2 = [2,6,2,5,1]
plt.bar(x,data2, color = 'g')
plt.show()
```

**数据被遮盖**

```python
# 多组数据

x = np.array([1,3,5,7,9])
data = np.array([6,4,2,6,2])
plt.bar(x,data)


x2 = x + 1
data2 = [2,6,2,5,1]
plt.bar(x2,data2, color = 'g')
plt.show()
```

```python
x = np.random.randint(1,30,size=10)

data1 = abs(x-np.mean(x))

data2 = x-np.mean(x)

plt.figure()

plt.scatter(x,data1,s=x*3,c="b",marker="*")

plt.scatter(x,data2,s=x[0]*5,c="r",marker="h")

plt.show()
```

```python
plt.figure(figsize=(10,5))

year=["2016","2017","2018","2019"]

x_A=[1,3,5,7]

score_A = [40,50,59,60]

w =0.5

plt.bar(x_A,score_A,width = w,label="A")

 

x_B=np.array(x_A)+w

score_B = [50,70,80,90]

plt.bar(x_B,score_B,width=w,label="B")

plt.xlabel("year")

plt.ylabel("score")

plt.legend()

plt.xticks(x_B-0.25,year)

plt.show()
```

## 直方图（histgram）
直方图形式类似于柱状图 ，通过将数据组合在一起来显示分布，比如：显示各年龄段的数据个数

plt.hist(x, bins, rwidth )

x 数据

rwidth 柱子宽度比例，默认为1

bins 分组的个数或分组边界，如4或[20, 30, 40, 50, 60]

除了最后一组，其余组的右边界是开放的
[20, 30), [30, 40), [40, 50), [50, 60]

```python
data  = np.random.randn(10000)
```

```python
plt.figure()
plt.hist(data, 40)
plt.show()
```

```python
bins = [-4,-3,-2,-1,0,1,2,3,4]
plt.hist(data,bins,rwidth= 0.9)
plt.show()
```

## 矩阵绘图
矩阵绘图通过可视化的方式呈现矩阵数据，可用于显示三维信息

plt.imshow(X, cmap)

X 矩阵数据（二维数组）

cmap 颜色主题
https://matplotlib.org/users/colormaps.html

plt.colorbar() 为图像添加颜色条

```python
data = np.random.rand(10,12)
type(data)
```

```python
plt.imshow(data)
plt.colorbar()
plt.show()
```

```python
plt.imshow(data, cmap = plt.cm.gnuplot)
plt.colorbar()
plt.show()
```

### ticks 设置

```python
data=np.random.randint(1,100,size=20).reshape(4,5)

plt.imshow(data,cmap=plt.cm.Blues)

plt.colorbar()

plt.xticks(range(data.shape[1]),["2014","2015","2016","2017","2018"])

plt.yticks(range(data.shape[0]),["January","February","March","April"])

plt.xlabel("year")

plt.ylabel("month") 
```

fmt 颜色 标记 线型的样式

如 r.-- 表示 红色 点标记 虚线
